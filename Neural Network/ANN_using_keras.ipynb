```python
from sklearn.datasets import load_iris
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical

data1 = load_iris()
a1 = data1.data
b1 = data1.target

# Convert target labels to one-hot encoding
b1 = to_categorical(b1) # This line is changed

model = Sequential()
model.add(Dense(12, input_shape=(4,), activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(3, activation='softmax')) # This line is changed
# compile the keras model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy']) # This line is changed
# fit the keras model on the dataset
model.fit(a1, b1, epochs=150, batch_size=10)
# evaluate the keras model
_, accuracy = model.evaluate(a1, b1)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.utils import to_categorical
from sklearn.preprocessing import StandardScaler


# 1. Data Preprocessing:
dataset = pd.read_csv('wine_data.csv')
p2 = dataset.iloc[:, 0:11].values
q3 = dataset.iloc[:, 11].values
q4 = to_categorical(q3)
p2_train, p2_test, q4_train, q4_test = train_test_split(p2, q4, test_size=0.2, random_state=42)

# Feature Scaling:
scaler = StandardScaler()
#Fit and transform only on the training features (p2_train)
p2_train = scaler.fit_transform(p2_train)
#Transform the test features (p2_test) using the fitted scaler
p2_test = scaler.transform(p2_test)

# 2. Model Architecture:
model = Sequential()
model.add(Dense(256, input_shape=(11,), activation='relu'))  # Increased neurons, added L2 regularization
model.add(Dense(128, activation='relu'))  # Added L2 regularization
model.add(Dense(64, activation='relu'))  # Changed activation function
model.add(Dense(32, activation='relu'))  # Added another layer
#Make sure the output layer has the correct number of classes (q4.shape[1])
model.add(Dense(q4.shape[1], activation='softmax'))

# 3. Compile:
model.compile(loss='categorical_crossentropy', optimizer='rmsprop', metrics=['accuracy'])  # Changed optimizer

# 5. Training:
model.fit(p2_train, q4_train, epochs=200, batch_size=32)  # Adjusted epochs, batch size, and added validation/early stopping

# 6. Evaluation:
_, accuracy = model.evaluate(p2_test, q4_test)
print('Accuracy: %.2f' % (accuracy*100))
```

```python
from numpy import loadtxt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from tensorflow.keras.utils import to_categorical
from sklearn.model_selection import train_test_split

dataset = loadtxt('wine_data (1).csv', delimiter=',', dtype=str, skiprows=1)

X = dataset[:, 0:12].astype(float)
y = dataset[:, 11]

encoder = LabelEncoder()
encoder.fit(y)
encoded_Y = encoder.transform(y)
num_classes = len(encoder.classes_)
dummy_y = to_categorical(encoded_Y, num_classes=num_classes)

X_train, X_test, y_train, y_test = train_test_split(X, dummy_y, test_size=0.2, random_state=42)

model = Sequential()
model.add(Dense(32, input_shape=(12,), activation='relu'))
model.add(Dense(16, activation='relu'))
model.add(Dense(num_classes, activation='softmax'))


model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])

history = model.fit(X_train, y_train, epochs=8, batch_size=25, validation_data=(X_train, y_train))
model.summary()


_, accuracy = model.evaluate(X_train, y_train)
print('Accuracy: %.2f' % (accuracy * 100))


import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Training', 'Validation'], loc='lower right')
plt.show()
```
